```
def soma(a, b):
    return a + b
def subtrair(a, b):
    return a - b
def multiplicar(a, b):
    return a * b
def dividir(a, b):
    return a / b

while True:
    print("\n" + "="*20)
    print("      MENU")
    print("="*20)
    print("1. Somar")
    print("2. Subtrair")
    print("3. Multiplicar")
    print("4. Dividir")
    print("5. Sair")
    print("="*20)

    opcao = input("Escolha uma opção (1-5): ")

    if opcao == '5':
        print("Saindo do programa... Até logo!")
        break

    if opcao in ['1', '2', '3', '4']:
        try:
            num1 = float(input("Digite o primeiro número: "))
            num2 = float(input("Digite o segundo número: "))
        except ValueError:
            print("digite apenas números válidos!")
            continue

        if opcao == '1':
            print(f"\nResultado: {num1} + {num2} = {soma(num1, num2)}")
        elif opcao == '2':
            print(f"\nResultado: {num1} - {num2} = {subtrair(num1, num2)}")
        elif opcao == '3':
            print(f"\nResultado: {num1} * {num2} = {multiplicar(num1, num2)}")
        elif opcao == '4':
            print(f"\nResultado: {dividir(num1, num2)}")

    else:
        print("Opção inválida! Tente novamente.")
```
