```
nota = float(input("Nota: "))
horas = float(input("Total de horas do curso: "))
faltas = float(input("Horas de falta: "))

# Calcula a frequência (ex: 100h e 10h de falta = 90% de presença)
presenca = ((horas - faltas) / horas) * 100

if nota >= 50 and presenca >= 75:
    print("Aprovado")
else:
    print("Reprovado")
```

```
nome = input("Nome: ")
escola = input("Escola (HRC ou outra): ").upper()
nota = float(input("Nota: "))
horas = float(input("Total de horas: "))
faltas = float(input("Faltas: "))

# Calcula a frequência em %
freq = ((horas - faltas) / horas) * 100

# Define o mínimo de presença (50 se for HRC, senão 40)
minimo = 50 if escola == "HRC" else 40

# Verifica primeiro a frequência
if freq < minimo:
    print(f"{nome}: Reprovado por falta")
# Se a frequência estiver ok, verifica a nota
elif nota > 70:
    print(f"{nome}: Aprovado")
elif nota >= 50:
    print(f"{nome}: Exame")
elif nota >= 30:
    print(f"{nome}: Recuperação")
else:
    print(f"{nome}: Reprovado por nota")
```
